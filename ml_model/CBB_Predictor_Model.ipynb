```python
import pandas as pd
```

```python
games = pd.read_csv('/content/cbb_games_23-26.csv')
```

```python
games["Date"] = pd.to_datetime(games["Date"])
games.dtypes
```

```python
all_teams = pd.concat([games['Team'], games['Opponent']]).unique()

# Create a mapping dictionary
team_to_code = {team: i for i, team in enumerate(sorted(all_teams))}

# Apply to both columns
games['Team_code'] = games['Team'].map(team_to_code)
games['Op_code'] = games['Opponent'].map(team_to_code)

usc_as_team = games[games['Team'] == 'southern-california']['Team_code'].iloc[0]
usc_as_opp = games[games['Opponent'] == 'southern-california']['Op_code'].iloc[0]
print(f"\nUSC code consistency check: {usc_as_team} == {usc_as_opp} ? {usc_as_team == usc_as_opp}")
```

```python
games.head()
```

```python
games["Day_code"] = games["Date"].dt.dayofweek
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier(n_estimators=50, min_samples_split=10, random_state=1)
```

```python
train = games[games["Date"] < '2025-01-01']
```

```python
test = games[games["Date"] >= '2025-01-01']
```

```python
predictors = ['Game_num','Day_code', 'Home_away','Team_code','Op_code','Team_pts','Opp_pts','Team_FG%','Team_3P%','Team_2P%','Team_eFG%',
            'Team_FT%','Team_ORB','Team_DRB','Team_TRB','Team_AST','Team_STL','Team_BLK','Team_TOV','Team_PF','Opponent_FG%','Opponent_3P%',
            'Opponent_2P%','Opponent_eFG%','Opponent_FT%','Opponent_ORB','Opponent_DRB','Opponent_TRB','Opponent_AST','Opponent_STL','Opponent_BLK',
            'Opponent_TOV','Opponent_PF']
```

```python
rf.fit(train[predictors], train["Won"])
```

```python
preds = rf.predict(test[predictors])
```

```python
from sklearn.metrics import accuracy_score
```

```python
acc = accuracy_score(test["Won"], preds)
```

```python
acc
```

```python
combined = pd.DataFrame(dict(actual=test["Won"], prediction=preds))
```

```python
pd.crosstab(index=combined["actual"], columns=combined["prediction"])
```

```python
win_probability = rf.predict_proba(test[predictors])[:,1]
print(f"Win probability: {win_probability[2]:.2%}")
```

```python
# Create a results dataframe
results = pd.DataFrame({
    'Team': test['Team'].values,
    'Opponent': test['Opponent'].values,
    'Home_Away': test['Home_away'].map({0: 'Away', 1: 'Home', 2: 'Neutral'}),
    'Predicted_Win_Prob': win_probability,
    'Actual_Result': test['Won'].values,
    'Correct': (win_probability > 0.5).astype(int) == test['Won'].values
})

# Look at first 10 games
print(results.head(10))
```
