```python
def add_rolling_stats_with_merge(df, window=3):
    """
    Simpler version: Calculate rolling stats for all teams
    """
    df = df.copy()

    stats = ['FG%', '3P%', '2P%', 'FT%', 'ORB', 'DRB',
             'AST', 'STL', 'BLK', 'TOV', 'pts']

    # Build a complete database of all teams' games
    all_team_games = []

    # Add Big 10 games (from Team perspective)
    for idx, row in df.iterrows():
        game_record = {
            'team': row['Team'],
            'date': row['Date']
        }
        for stat in stats:
            game_record[stat] = row[f'Team_{stat}']
        all_team_games.append(game_record)

    # Add opponent games (from Opponent perspective)
    for idx, row in df.iterrows():
        game_record = {
            'team': row['Opponent'],
            'date': row['Date']
        }
        for stat in stats:
            if stat == 'pts': # Special handling for 'pts' stat
                game_record[stat] = row['Opp_pts']
            else:
                game_record[stat] = row[f'Opponent_{stat}']
        all_team_games.append(game_record)

    # Create unified dataframe
    all_games_df = pd.DataFrame(all_team_games)
    all_games_df = all_games_df.drop_duplicates(subset=['team', 'date'])
    all_games_df = all_games_df.sort_values(['team', 'date'])

    # Calculate rolling stats
    for stat in stats:
        all_games_df[f'{stat}_rolling'] = all_games_df.groupby('team')[stat].transform(
            lambda x: x.shift(1).rolling(window=window, min_periods=1).mean().round(3)
        )
        all_games_df[f'{stat}_rolling'] = all_games_df[f'{stat}_rolling'].fillna(all_games_df[stat])

    # Merge back to original df for Team
    team_rolling = all_games_df[['team', 'date'] +
                                [f'{stat}_rolling' for stat in stats]].copy()
    team_rolling = team_rolling.rename(columns={
        'team': 'Team',
        'date': 'Date',
        **{f'{stat}_rolling': f'Team_{stat}_L{window}' for stat in stats}
    })
    df = df.merge(team_rolling, on=['Team', 'Date'], how='left')

    # Merge back for Opponent
    opp_rolling = all_games_df[['team', 'date'] +
                               [f'{stat}_rolling' for stat in stats]].copy()
    opp_rolling = opp_rolling.rename(columns={
        'team': 'Opponent',
        'date': 'Date',
        **{f'{stat}_rolling': f'Opp_{stat}_L{window}' for stat in stats}
    })
    df = df.merge(opp_rolling, on=['Opponent', 'Date'], how='left')

    print(f"âœ“ Rolling stats for all teams: {len(df)} rows")

    return df

# Usage
df = add_rolling_stats_with_merge(games, window=3)

print("âœ“ Rolling stats added correctly")
```

```python
def comprehensive_verification(df):
    """
    Comprehensive check that stats are aligned correctly
    """
    print("=" * 60)
    print("VERIFICATION: Rolling Stats Alignment")
    print("=" * 60)

    # Find a game where both teams have data
    test_game = df[
        (df['Team'] == 'southern-california') &
        (df['Opponent'] == 'chattanooga')
    ].iloc[0]

    date = test_game['Date']

    # Find the reverse game (UCLA's perspective)
    reverse_game = df[
        (df['Team'] == 'ucla') &
        (df['Opponent'] == 'southern-california') &
        (df['Date'] == date)
    ].iloc[0]

    print(f"\nGame Date: {date}")
    print(f"\nUSC's Row:")
    print(f"  Team: {test_game['Team']}")
    print(f"  Opponent: {test_game['Opponent']}")
    print(f"  Team_PTS_L3: {test_game.get('Team_pts_L3', 'N/A')}")
    print(f"  Opp_PTS_L3: {test_game.get('Opp_pts_L3', 'N/A')}")

    print(f"\nUCLA's Row (opposite perspective):")
    print(f"  Team: {reverse_game['Team']}")
    print(f"  Opponent: {reverse_game['Opponent']}")
    print(f"  Team_PTS_L3: {reverse_game.get('Team_pts_L3', 'N/A')}")
    print(f"  Opp_PTS_L3: {reverse_game.get('Opp_pts_L3', 'N/A')}")

    # The critical check
    usc_opp_rolling = test_game.get('Opp_pts_L3')
    ucla_team_rolling = reverse_game.get('Team_pts_L3')

    print(f"\n{'='*60}")
    print("CRITICAL CHECK:")
    print(f"USC's Opponent Rolling (should be UCLA's team rolling): {usc_opp_rolling}")
    print(f"UCLA's Team Rolling: {ucla_team_rolling}")
    print(f"Match: {abs(usc_opp_rolling - ucla_team_rolling) < 0.01 if pd.notna(usc_opp_rolling) and pd.notna(ucla_team_rolling) else 'N/A'}")
    print(f"{'='*60}")

    return abs(usc_opp_rolling - ucla_team_rolling) < 0.01 if pd.notna(usc_opp_rolling) and pd.notna(ucla_team_rolling) else False

# Run verification
is_correct = comprehensive_verification(df)
if is_correct:
    print("\nâœ“ Rolling stats are correctly aligned!")
else:
    print("\nâœ— Warning: Rolling stats may not be aligned correctly")
```
